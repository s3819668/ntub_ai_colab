```python
import pandas as pd
df=pd.read_csv("pima2.csv")
df.head()
```

```python
df.columns
```

```python
data_X=df[['pregnancies', 'glucose', 'bloodPressure', 'skinThickness', 'insulin',
       'BMI', 'diabetesPedigree', 'age']]
data_y=df.outcome
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
train_X,test_X,train_y,test_y=train_test_split(data_X,data_y,test_size=0.3,random_state=1)
train_X.info()
clf=LogisticRegression()
clf.fit(train_X,train_y)
p=clf.predict(test_X)
print(sum(p==test_y)/len(p))
```

評估指標.png

```python
TP=sum((p==test_y)& (test_y==1))
TN=sum((p==test_y)& (test_y==0))
FP=sum((p!=test_y)& (test_y==1))
FN=sum((p!=test_y)& (test_y==0))
print(TP,TN,FP,FN)

accuracy=(TP+TN)/(TP+TN+FP+FN)
print("準確率",accuracy)

precision=TP/(TP+FP)
print("精確率",precision)

recall=TP/(TP+FN)
print("召回率",recall)
f1=2/(1/precision+1/recall)
print("f1",f1)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

print(accuracy_score(test_y,p))
print(precision_score(test_y,p))
print(recall_score(test_y,p))
print(f1_score(test_y,p))
```
