```python
import pandas as pd
df=pd.read_csv("StudentsPerformance-3.csv")
df.replace(['male','female'],[1,2],inplace=True)
df.replace(['group A','group B','group C','group D','group E'],[1,2,3,4,5],inplace=True)
df.replace(["bachelor's degree", 'some college', "master's degree",
       "associate's degree", 'high school', 'some high school'],[1,2,3,4,5,6],inplace=True)
df.replace(['standard','free/reduced'],[1,2],inplace=True)
df.replace(['none', 'completed'],[1,2] ,inplace=True)

```

```python
df.columns
```

```python
data_X=df.drop(['Unnamed: 0', 'math', 'reading', 'writing', 'mathPass'],axis=1)
```

```python
data_y=df.mathPass
```

```python
data_X
```

```python
data_y
```

```python
from sklearn.model_selection import train_test_split

train_X,test_X,train_y,test_y=train_test_split(data_X,data_y)
```

```python
train_X
```

```python
train_y
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()#建立空模型
clf.fit(train_X,train_y)
```

```python
p=clf.predict(test_X)
```
