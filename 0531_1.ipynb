```python
import numpy as np
IN=np.array([0.05,0.1])
W1=np.array([[0.15,0.25],[0.2,0.3]])
B1=np.array([0.35,0.35])

H1=np.dot(IN,W1)+B1
print(H1)
```

```python
Y1=1/(1+np.exp(-H1))
Y1
```

```python
W2=np.array([[0.4,0.5],[0.45,0.55]])
B2=np.array([0.6,0.6])

OUT=np.dot(Y1,W2)+B2
print(OUT)
Y2=np.exp(OUT)/sum(np.exp(OUT))
Y2

T=np.array([0,1])
L=sum((Y2-T)**2)/2
L
```

```python
a=-(T[0]-Y2[0])

b=Y2[0]*(1-Y2[0])

c=Y1[0]
print(a,b,c)

w=0.4
lr=0.5
w=w-lr*(a*b*c)
print(w)
```
