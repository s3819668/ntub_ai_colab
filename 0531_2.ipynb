```python
import pandas as pd
df=pd.read_csv("seed.csv")
df.head()
```

```python
df.species.value_counts()
df.columns
```

```python
data_X=df[['area', 'perimeter', 'compactness', 'kernelLength', 'kernelWidth',
       'asymmetry', 'kernelGroove']]
data_y=df.species
```

```python
data_X
```

```python
from sklearn.model_selection import train_test_split
train_X,testX,train_y,test_y=train_test_split(data_X,data_y)
```

```python
from sklearn.neural_network import MLPClassifier

clf=MLPClassifier()
clf.fit(train_X,train_y)
```

```python
p=clf.predict(testX)
print(sum(p==test_y)/len(p))
```
